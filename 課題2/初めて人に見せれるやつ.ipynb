```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# クラスラベルと整数を対応させるディクショナリを生成
embarked_mapping = {'C': 0, 'S': 1, 'Q': 2}
embarked_mapping
sex_mapping = {'male': 0, 'female': 1}
sex_mapping
```

```python
# マッピングの実行
train['Embarked'] = train['Embarked'].map(embarked_mapping)
train['Sex'] = train['Sex'].map(sex_mapping)
train
```

Embarkedについては、mappingよりencodingの方がいい。3つに分けた方は0.05くらい相関係数上がる。

```python
embarked_dm = pd.get_dummies(train['Embarked'])
train = pd.concat((train,embarked_dm),axis=1)
train = train.drop('Embarked',axis=1)
```

```python
train
```

そっか名前が0,1,2になるなら上書きしてやるのは間違いでそもそもやらない方がいい

```python
#元からやるのも面倒なので,pandas.DataFrameの行名、列名の変更をする
train_new = train.rename(columns={0: 'C',1:'S',2:'Q'})
```

```python
train_new
```

```python
#それ以外は相関係数を見てみる
train_new.corr()
```

相関係数が低い、Passengerld,Age,SibSp,Parch,Qは要らなそう

```python
import seaborn as sns
#一応相関係数のヒートマップも見ておく
# ヒートマップの表示
plt.figure(figsize=(12, 9))
# annot : 数値を表示するかどうか(annotation)
# square: 四角を正方形に合わせるかどうか
# fmt   : 表示する数値の形式(formatting)
sns.heatmap(train_new.corr(), annot=True, square=True, fmt='.2f')
plt.show()
```

```python
#相関係数が低いやつは思い切って、カラム消去+扱いにくいNameとTicket＋欠損値が多すぎるCabinもカラム消去
train_new = train_new.drop(['PassengerId','Age','SibSp','Parch','Q','Name','Ticket','Cabin'], axis=1)
```

```python
train_new
```

## 学習

```python
train_labels = train_new['Survived'].values
train_features = train_new
train_features
```

```python
train_features.drop('Survived',axis=1,inplace=True)
train_features = train_features.values
train_features
```

```python
#SVMで学習
from sklearn import svm
svm = svm.LinearSVC()
svm.fit(train_features,train_labels)
```

## テスト

```python
test = pd.read_csv('test1.csv')
```

```python
#pandasでone-hotエンコーディングを使い、ダミー変数を作成
sex_dm = pd.get_dummies(test['Sex'])
embarked_dm = pd.get_dummies(test['Embarked'])
#元のデータに連結する（その際、元となったデータは消す）
test = pd.concat((test,sex_dm),axis=1)
test = test.drop('Sex',axis=1)
test = test.drop('female',axis=1)
#sexはmaleの0,1で判断できるので、femaleは消す

test = pd.concat((test,embarked_dm),axis=1)
test = test.drop('Embarked',axis=1)

test_new = test.drop(['PassengerId','Age','SibSp','Parch','Q','Name','Ticket','Cabin'], axis=1)
```

```python
test_features = test_new.values.astype(np.int64)
test_features
```

```python
#各点を分類する
y_test_pred = svm.predict(test_features)
```

```python
df_out = pd.read_csv('test1.csv')
df_out["Survived"] = y_test_pred
```

```python
df_out[["PassengerId","Survived"]].to_csv('submission.csv')
```
