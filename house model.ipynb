```python
# trainデータを用意
# DataFrameの.valuesメソッドでarrayに変換できます
X = train_df.loc[:,['1stFlrSF','GarageArea','GrLivArea','TotalBsmtSF']]
```

```python
X
```

```python
#　testデータを用意
#DataFrameの.valuesメソッドでarrayに変換できます
y = train_df.loc[:,['SalePrice']].values
```

```python
#どのように入っているか、最初の10個を表示して確認
y[:10]
```

```python
#ホールド・アウト法によるデータの分割
#(X,y,テストにどの割合を割くか、分裂のしかたを決める乱数生成機)
#テストデータの割合は、test_size引数に指定することで変更可能です(今回は、train:test = 7:3)
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=0)
```

```python
#データのサンプル数を確認
print(X.shape)
print(y.shape)
```

```python
# trainデータのサンプル数を確認
print(X_train.shape)
print(y_train.shape)
```

```python
#testデータのサンプル数を確認
print(X_test.shape)
print(y_test.shape)
```

```python
# 重回帰の線形回帰を実行
from sklearn.linear_model import LinearRegression
Ir = LinearRegression()
```

```python
#　重回帰線形モデルの学習
# trainデータのみを用いていることに注意
Ir.fit(X_train,y_train)
```

```python
# 定数項を出力
Ir.intercept_
```

```python
#偏回帰係数を出力
#[LSTATの係数、RMの係数]
Ir.coef_
```

```python
print('モデルの性能評価')
```

```python
#決定係数をtrainとtestに分けて出力
#　過学習しているかどうかを確認
print('R^2')
print('train: %.3f' % Ir.score(X_train,y_train))
print('test: %.3f' % Ir.score(X_test,y_test))
```

```python
# 自由度調整済み決定係数
# (決定係数, trainまたはtestのサンプル数, 利用した特徴量の数)
def adjusted(score, n_sample, n_features):
    adjusted_score = 1 - (1 - score) * ((n_sample - 1) / (n_sample - n_features - 1))
    return adjusted_score
```

```python
# 自由度調整済み決定係数を出力
print('adjusted R^2')
print('train: %3f' % adjusted(Ir.score(X_train, y_train), len(y_train), 2))
print('test : %3f' % adjusted(Ir.score(X_test, y_test), len(y_test), 2))
```

```python
print('全然ダメなモデルですね')
```

```python
# MSEを出力する関数を読み込む
from sklearn.metrics import mean_squared_error as mse

# RMSEをtrainとtestに分けて出力
# 過学習をしているかどうかを確認
print('RMSE')
print('train: %.3f' % (mse(y_train, Ir.predict(X_train)) ** 1/2))
print('test : %.3f' % (mse(y_test, Ir.predict(X_test)) ** 1/2))
```
